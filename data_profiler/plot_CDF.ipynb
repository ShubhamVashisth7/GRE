```python
import numpy as np
import matplotlib.pyplot as plt

dataset = 'planet'
with open(f'../datasets/{dataset}', 'rb') as f:
    data = np.fromfile(f, dtype=np.uint64)[0:]
    

data_sorted = np.sort(data)
y = np.arange(1, len(data_sorted) + 1) / len(data_sorted)

plt.figure(figsize=(6, 2.5), dpi=300)
fontsize = 8

plt.subplot(1, 2, 1)
plt.plot(data_sorted, y, color='cornflowerblue', linewidth=1.5)
plt.yticks(np.arange(0.0, 1.1, 0.25), ['0%', '25%', '50%', '75%', '100%'], fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.ylim(top=1)
plt.xlim(left=0.0) 
plt.title(dataset, fontsize=fontsize+4)
plt.xlabel('Key domain', fontsize=fontsize)

dataset = 'genome'
with open(f'../datasets/{dataset}', 'rb') as f:
    data = np.fromfile(f, dtype=np.uint64)[0:]

data_sorted = np.sort(data)
y = np.arange(1, len(data_sorted) + 1) / len(data_sorted)

plt.subplot(1, 2, 2)
plt.plot(data_sorted, y, color='cornflowerblue', linewidth=1.5)
plt.yticks(np.arange(0.0, 1.1, 0.25), ['0%', '25%', '50%', '75%', '100%'], fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.ylim(top=1)
plt.title(dataset, fontsize=fontsize+4)
plt.xlabel('Key domain', fontsize=fontsize)
plt.xlim(left=0.0) 

plt.tight_layout()
plt.savefig('out/fig1_CDF.jpg', bbox_inches='tight', dpi=300)
plt.show()
```

```python
datasets = ['covid', 'libio', 'genome', 'osm']

plt.figure(figsize=(6, 5), dpi=300)
fontsize = 8

for i, dataset in enumerate(datasets, 1):
    with open(f'../datasets/{dataset}', 'rb') as f:
        data = np.fromfile(f, dtype=np.uint64)[0:]

    data_sorted = np.sort(data)
    y = np.arange(1, len(data_sorted) + 1) / len(data_sorted)

    plt.subplot(2, 2, i)
    plt.plot(data_sorted, y, color='cornflowerblue', linewidth=1.5)
    plt.yticks(np.arange(0.0, 1.1, 0.25), ['0%', '25%', '50%', '75%', '100%'], fontsize=fontsize)
    plt.xticks(fontsize=fontsize)
    plt.ylim(top=1)
    plt.xlim(left=0.0)
    plt.title(dataset, fontsize=fontsize+4)
    plt.xlabel('Key domain', fontsize=fontsize)

plt.tight_layout()
plt.savefig('out/CDF.png', bbox_inches='tight', dpi=300)
plt.show()
```
